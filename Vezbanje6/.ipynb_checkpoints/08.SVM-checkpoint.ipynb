```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
X, Y = load_breast_cancer(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42, stratify=Y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = [
    {
        'kernel': ['linear'],
        'C': [0.01, 0.1, 1, 10]
    },
    {
        'kernel': ['rbf'],
        'C': [0.01, 0.1, 1, 10],
        'gamma': [0.01, 0.1, 1, 10]
    }
]
```

```python
estimator = GridSearchCV(SVC(), param_grid=params, cv=10, verbose=25)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
model = estimator.best_estimator_
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

```python
num_instances = 1000
X = np.random.uniform(low=-1, high=1, size=(num_instances, 2))
```

```python
def f(x, y):
    return x**2 + y**2 <= 0.5
```

```python
y = f(X[:,0], X[:,1])
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
model = SVC(kernel='linear')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
model.fit(X_train, y_train)
```

```python
accuracy_score(y_train, model.predict(X_train))
```

```python
accuracy_score(y_test, model.predict(X_test))
```

```python
f1_score(y_train, model.predict(X_train))
```

```python
f1_score(y_test, model.predict(X_test))
```

```python
model = SVC(kernel='rbf')
```

```python
model.fit(X_train, y_train)
```

```python
accuracy_score(y_train, model.predict(X_train))
```

```python
accuracy_score(y_test, model.predict(X_test))
```

```python
print(f'Broj potpornih vektora: {model.n_support_}')
```

```python
print(f'Indeksi instanci koji su potporni vektori: {model.support_}')
```

```python
print(f'Sami potporni vekotir: {model.support_vectors_}')
```
