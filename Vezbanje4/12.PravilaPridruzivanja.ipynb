```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('../datasets/izborni_predmeti.xlsx')
```

```python
f = lambda x: x['NAZIV_SMERA'] in ['Informatika', 'Racunarstvo i informatika']
```

```python
data = data[data.apply(f, axis=1)]
```

```python
np.unique(data['NAZIV_SMERA'])
```

```python
# grupisanje po indeksu
data = data.groupby(['INDEKS'])['NAZIV_PREDMETA'].apply(list).values.tolist()
```
