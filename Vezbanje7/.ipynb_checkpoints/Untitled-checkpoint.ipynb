```python
from sklearn.decomposition import PCA
```

```python
def visualize(X, X_resampled, Y, Y_resampled, x_oppacity=0.5):
    pca = PCA(n_components=2)
    pca.fit(X)
    
    X_pca = pca.transform(X)
    X_resampled_pca = pca.transform(X_resampled)
    
    X_stacked = [X_pca, X_resampled_pca]
    Y_stacked = [Y, Y_resampled]
    
    plt.figure(figsize=(20, 5))
    
    for i in range(2):
        ax = plt.subplot(1, 2, i+1)
        color = ['b' if y==-1 else 'r' for y in Y_stacked[i]]
        alpha = [x_oppacity if y==-1 else 1-x_oppacity for y in Y_stacked[i]]
        plt.scatter(X_stacked[i][:,0], X_stacked[i][:,1], color=color, alpha=alpha)
        plt.title(f'{"After" if i else "Before"}', color='blue' if i else 'green')
        
    plt.show()
```

```python
data = pd.read_csv('../datasets/dogs.csv')
```

```python
data
```

```python
feature_names = data.columns[1:]
```

```python
X = data[feature_names]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X)
```

```python
X = pd.DataFrame(scaler.transform(X), columns=feature_names)
```

```python
X 
```

```python
plt.scatter(X['height'], X['weight'])
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3, n_init='auto')
```

```python
km.fit(X)
```

```python
km.labels_
```

```python
km.cluster_centers_
```

```python
centroids = pd.DataFrame(km.cluster_centers_, columns=feature_names)
```

```python
centroids
```

```python
plt.scatter(centroids['height'], centroids['weight'], marker='X', label='centroids')

for c in np.unique(km.labels_):
    elems = X[km.labels_ == c]
    plt.scatter(elems['height'], elems['weight'], label=c)
    
plt.title('KMeans with 3 clusters')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z = linkage(X, method='single')
```

```python
X = X.set_index(data['breed'])
```

```python
X
```

```python
_ = dendrogram(Z, labels=X.index, leaf_rotation=90)
```

```python
Z = linkage(X, method='average')
```

```python
_ = dendrogram(Z, labels=X.index, leaf_rotation=90)
```

```python
Z = linkage(X, method='complete')
```

```python
_ = dendrogram(Z, labels=X.index, leaf_rotation=90)
```

```python
ks = 3
linkages = ['single', 'average', 'complete']

plt.figure(figsize=(20, 5))
idx = 1

for linkage in linkages:
    model = AgglomerativeClustering(n_clusters=ks, linkage=linkage)
    model.fit(X)
    
    plt.subplot(1, len(linkages), idx)
    idx += 1
    
    for c in np.unique(model.labels_):
        elems = X[model.labels_ == c]
        plt.scatter(elems['height'], elems['weight'])
        
    
```
